```python
import csv
import pandas as pd
from scipy.stats.stats import pearsonr

responses = []

with open('/processed_data/strategy_coded_runs.csv', 'r') as f:
    csvr = csv.reader(f, delimiter=',')
    
    run_id = 0
    index = 0
    
    for i, row in enumerate(csvr):
        if i == 0:
            continue
        
        if all([len(i) == 0 for i in row]):
            run_id += 1
            index = 0
        else:
            responses.append((run_id, row[0], row[1], row[2], row[3], row[4], ','.join([i for i in row[5:] if len(i) > 0]), index))
            index += 1

df = pd.DataFrame({'run_id': pd.Series([r[0] for r in responses]),
                   'scoping': pd.Categorical.from_array([r[1] for r in responses]),
                   'riff_type': pd.Categorical.from_array([r[2] for r in responses]),
                   'pair': pd.Categorical.from_array([r[3] for r in responses]),
                   'reach_back': pd.Categorical.from_array([r[4] for r in responses]),
                   'partial': pd.Categorical.from_array([r[5] for r in responses]),
                   'answer': pd.Series([r[6] for r in responses]),
                   'answer_index': pd.Series(r[7] for r in responses)
                   })

stat_cols = ['scoping', 'riff_type', 'pair', 'reach_back', 'partial']
```
