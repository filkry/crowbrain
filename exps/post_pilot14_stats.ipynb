```python
base_data_dirs = ['/processed_data/pilot14',
                  '/processed_data/pilot13',
                  '/processed_data/pilot12',
                  '/processed_data/pilot11',]

originality_csvs = map(lambda x: '/processed_data/pilot14_metrics/' + x + '_originality.csv',
                         ['charity', 'iPod', 'turk', 'forgot_name'])

manual_csvs = map(lambda x: '/processed_data/pilot14_metrics/' + x + '-scores.csv',
                         ['fil', 'mike'])

heirarchical_cluster_csvs = map(lambda x: '/processed_data/pilot14_metrics/' + x,
                         ['h_clusters.csv' , 'h_clusters_map.csv'])

nwf = lambda x: '/processed_data/pilot14_metrics/' + x + '-newick.csv'
output_newick = dict();
for key in ['charity', 'iPod', 'turk', 'forgot_name']:
    output_newick[key] = nwf(key)

# Read in a bunch of base data

def read_files(fils):
    rows = []
    for f in fils:
        rows = rows + read_file(f)
    return rows
     

def read_base_data(dirs):
    rows = []
    for d in dirs:
      for f in os.listdir(d):
        full_name = os.path.join(d,f)
        if os.path.isdir(full_name):
            print "I deleted the code here"
        else:
          if f == 'answers.csv':
            rows = rows + read_file(full_name)
    return rows

def read_file(f):
    with open(f) as fin:
        l = fin.readline()
        # If we see the follow, it's an early run we can ignore
        if 'Number of Answers' in l:
            return
        fin.seek(0)
        sep = '|'
        # Guess separator character
        if l.count(',') > l.count('|'):
            sep = ','
            
        # Just cache for now
        rows = []
        for row in csv.reader(fin, delimiter=sep):
            if 'hashed_worker_id' in row or 'worker_id' in row or 'cluster_parent' in row:
                continue
            rows.append(row)
            
        return rows

def clean_missing(v):
    return 0 if v == '' else v

def series_from_row(rows, index, t):
    if t == object:
        return pd.Series([row[index] for row in rows], dtype=t)
    else:
        return pd.Series([clean_missing(row[index]) for row in rows], dtype=t)

merge_column_names = ['hashed_worker_id', 'question_code', 'answer_num', 'num_requested']
    
# Read in the main processed data files
all_rows = read_base_data(base_data_dirs)
num_responses = len(all_rows)

print "num base responses:", num_responses

df_base = pd.DataFrame({'hashed_worker_id': pd.Series([row[0] for row in all_rows], dtype=object),
            'question_code': pd.Series([row[2] for row in all_rows], dtype=object),
            'num_requested': pd.Series([row[5] for row in all_rows], dtype=uint8),
            'answer_num': pd.Series([row[6] for row in all_rows], dtype=uint8),
            'answer': pd.Series([row[7] for row in all_rows], dtype=object),
            'answer_word_count': pd.Series([row[8] for row in all_rows], dtype=uint32),
})

# Read in heirarchical clusters
rows = read_file(heirarchical_cluster_csvs[0])

df = pd.merge(df_base,
              pd.DataFrame({'hashed_worker_id': series_from_row(rows, 6, object),
                  'question_code': series_from_row(rows, 0, object),
                  'num_requested': series_from_row(rows, 8, uint8),
                  'answer_num': series_from_row(rows, 5, uint8),
                  'h_cluster': series_from_row(rows, 1, uint16),
                  'h_cluster_valid': pd.Series([1 for row in rows], dtype=uint8),}),
              'outer', merge_column_names)

# Read in manual codes

def read_manual_csv(name):
    index = 0 if name == 'fil' else 1
    rows = read_file(manual_csvs[index])
    
    df = pd.DataFrame({'hashed_worker_id': series_from_row(rows, 0, object),
            'question_code': series_from_row(rows, 2, object),
            'num_requested': series_from_row(rows, 5, uint8),
            'answer_num': series_from_row(rows, 6, uint8),
            ('originality_%s' % name): series_from_row(rows, 12, uint8),
            ('utility_%s' % name): series_from_row(rows, 13, uint8),
            ('realistic_%s' % name): series_from_row(rows, 14, uint8),
            ('underdefined_%s' % name): series_from_row(rows, 15, uint8),
            ('distance_%s' % name): series_from_row(rows, 16, uint8),
            ('%s_valid' % name): pd.Series([1 for row in rows], dtype=uint8),
    })
    return df

df = pd.merge(df, read_manual_csv('mike'), 'outer', merge_column_names)
df = pd.merge(df, read_manual_csv('fil'), 'outer', merge_column_names)

# Read in exterior computeted originality (ISS) scores

all_rows = read_files(originality_csvs)

df = pd.merge(df,
              pd.DataFrame({'hashed_worker_id': series_from_row(all_rows, 4, object),
                            'question_code': series_from_row(all_rows, 7, object),
                            'answer_num': series_from_row(all_rows, 3, uint8),
                            'num_requested': series_from_row(all_rows, 6, uint8),
                            'originality_inverse_sum_similarity': series_from_row(all_rows, 2, float64),
                            'exterior_originality_valid': pd.Series([1 for row in all_rows], dtype=uint8),}),
              'outer', merge_column_names)

print "Dropping unlimited condition"
df = df[~(df['num_requested'] == 0)]

print "num merged responses:", len(df)
```

# Generate some extra columns of metrics

+ h_cluster_high

+ h_cluster_high_o_score

```python
def dumb_strip(s):
    return ''.join([c for c in s if c in "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz1234567890 "])
```

```python
# Compute and store a map of the heirarchical clustering

h_cluster_map = {qc: nx.DiGraph() for qc in question_codes}
h_cluster_label = {qc: dict() for qc in question_codes}

rows = read_file(heirarchical_cluster_csvs[1])
for qc, child, parent, label in rows:
    if parent != '':
        h_cluster_map[qc].add_edge(int(parent), int(child))
    else:
        h_cluster_map[qc].add_node(int(child))
    if len(label) > 0:
        h_cluster_label[qc][int(child)] = dumb_strip(label)

# Add a column for "high level" clusters (any idea in the same subgraph of the cluster map)

h_clustered = df[(df['h_cluster_valid'] > 0)]

U = {qc: nx.Graph(h_cluster_map[qc]) for qc in question_codes}
h_cluster_highs = {qc: nx.connected_component_subgraphs(U[qc]) for qc in question_codes}

# map low level clusters to high level clusters
cluster_lookup = {qc: np.zeros(max(h_clustered['h_cluster']) + 1) for qc in question_codes}

for qc in question_codes:
    sub_df = h_clustered[(h_clustered['question_code'] == qc)]
    for low in sub_df['h_cluster']:
        if low:
            for i, high in enumerate(h_cluster_highs[qc]):
                if low in high:
                    cluster_lookup[qc][low] = i
                    break

for qc in question_codes:
    sub_df = h_clustered[(h_clustered['question_code'] == qc)]
    if len(sub_df) == 0:
        continue
    
    qc_h_clus = sub_df['h_cluster']
    df['h_cluster_high'] = pd.Series([cluster_lookup[qc][x] for x in qc_h_clus],
                                     index=qc_h_clus.index)

    ## compute o-scores for these cluster types

    count = np.zeros(max(sub_df['h_cluster']) + 1)
    for c in qc_h_clus:
        count[c] += 1
    p_h_cluster = count / np.sum(count)
    df['h_cluster_o_score'] = pd.Series([1 - p_h_cluster[x] for x in qc_h_clus],
                                        index=qc_h_clus.index)

    count = np.zeros(len(h_cluster_highs[qc]))
    for i, c in enumerate(h_cluster_highs[qc]):
        count[i] += len(c)
    p_h_cluster_high = count / np.sum(count)
    s = pd.Series([1 - p_h_cluster_high[cluster_lookup[qc][x]] for x in qc_h_clus],
                  index=qc_h_clus.index)
    df['h_cluster_high_o_score'] = s
```

```python
def run_count(run, pass_func):
    count = 0
    for i in run.iterrows():
        row = i[1]
        if pass_func(row):
            count += 1
    return count

def run_mean(run, value_func):
    t = 0.0
    for i in run.iterrows():
        row = i[1]
        t += value_func(row)
    return t / len(run)

runs = df.groupby(['num_requested', 'hashed_worker_id', 'question_code'])

wids = pd.Series([wid for ((nr, wid, qc), run) in runs], dtype=object)
qcs = pd.Series([qc for ((nr, wid, qc), run) in runs], dtype=object)
nrs = pd.Series([nr for ((nr, wid, qc), run) in runs], dtype=uint16)

mwc_val = lambda x: x['answer_word_count']
mwc = pd.Series([run_mean(run, mwc_val) for ((nr, wid, qc), run) in runs],
                dtype=float64)

nmo_pass = lambda x: x['originality_mike'] > 0 and \
                     x['originality_fil'] > 0 and \
                     x['distance_mike'] == 0 and \
                     x['distance_fil'] == 0
nmo = pd.Series([run_count(run, nmo_pass) for ((nr, wid, qc), run) in runs],
                dtype=uint16)

nmu_pass = lambda x: x['utility_mike'] > 0 and \
                     x['utility_fil'] > 0
nmu = pd.Series([run_count(run, nmu_pass) for ((nr, wid, qc), run) in runs],
                dtype=uint16)

nmr_pass = lambda x: x['realistic_mike'] > 0 and \
                     x['realistic_fil'] > 0
nmr = pd.Series([run_count(run, nmr_pass) for ((nr, wid, qc), run) in runs],
                dtype=uint16)

nmun_pass = lambda x: x['underdefined_mike'] > 0 and \
                     x['underdefined_fil'] > 0
nmun = pd.Series([run_count(run, nmun_pass) for ((nr, wid, qc), run) in runs],
                dtype=uint16)
```

# Output newick-format clustering for outside visualization

```python
# TODO: need to do this separately for each idea code

for qc in question_codes:
    sub_df = df[(df['question_code'] == qc) & (df['h_cluster_valid'] > 0)]
    if len(sub_df) == 0:
        continue
    
    hcm = h_cluster_map[qc]
    hcl = h_cluster_label[qc]
    
    roots = []
    for n in hcm:
        if len(hcm.predecessors(n)) == 0:
            roots.append(n)
    
    # TODO: this function should be preprocessed, but our tree traversal and selection is fast
    def num_below(n):
        num = len(sub_df[(sub_df['h_cluster'] == n)])
        for succ in hcm.successors(n):
            num += num_below(succ)
        return num
            
    def tree_string(n):
        
        children = []
        
        c = sub_df[(sub_df['h_cluster'] == n)]
        for a in c['answer']:
            fixed = dumb_strip(a)
            children.append(fixed)
    
        for succ in hcm.successors(n):
            recur = tree_string(succ)
            if not recur is None:
                children.append(tree_string(succ))
            
        if len(children) > 0:
            label = hcl[n] if n in hcl else ""
            nb = num_below(n)
            pb = "%f" % (float(nb)/ len(sub_df) * 100)
            return "(" + ','.join(children) + ")" + label + " --count-" + str(nb) + " --percent-" + pb
        else:
            return None
        
    of = output_newick[qc]
    with open(of, 'w') as f:
        s = "("
        for i, r in enumerate(roots):
            ts = tree_string(r)
            if not ts is None:
                s += ts
                if i != len(roots) - 1:
                    s += ','
        s += ");"
        f.write(s + '\n')
```

# Response Measures

Binary measures on ideas, and summaries of these measures on runs. Each measure returns a pandas list of indices that pass the test.

```python
def bin_man_orig(f):
    original= f[(f['originality_mike'] > 0) & (f['originality_fil'] > 0)]
    return original.index

def bin_man_orig_noriff(f):
    original= f[(f['originality_mike'] > 0) & (f['originality_fil'] > 0) & (f['distance_mike'] == 0) & (f['distance_fil'] == 0)]
    return original.index

def bin_inmix(f):
    inmixed = set()
    
    for i in f.index:
        j = i
        
        while(f['distance_fil'][j] > 0):
            j = j - f['distance_fil'][j]
            if not j in f.index:
                j = -1
                break
        if i != j and j != -1:
            inmixed.add(j)
        
        j = i
        while(f['distance_mike'][j] > 0):
            j = j - f['distance_mike'][j]
            if not j in f.index: # handle bugs where distances go to other runs or we have rows missing codes completely
                j = -1
                break
        if i != j and j != -1:
            inmixed.add(j)
            
    return list(inmixed)

def bin_outmix(f):
    outmix = f[(f['distance_fil'] > 0) | (f['distance_mike'] > 0)] # remixed if either Mike or I gave it nonzero distance
    return outmix.index

def scalar_by_nr(dataframe, field, qc = None):
    f = dataframe if qc is None else dataframe[(dataframe['question_code'] == qc)]
    out = defaultdict(list)
    
    for i, row in f.iterrows():
        out[row['num_requested']].append(row[field])
    return out

def coded_only_df():
    return df[(df['mike_valid'] > 0) & (df['fil_valid'] > 0)]

def bin_pass_counts_by_nr(pass_func, coded_only, qc = None):
    base = coded_only_df() if coded_only else df
    f = base if qc is None else base[(base['question_code'] == qc)]
    f_runs = f.groupby(['num_requested', 'hashed_worker_id', 'question_code'])
    
    out = defaultdict(list)
    
    for (num_requested, wid, qc), run in f_runs:
        pass_ids = pass_func(run)
        out[num_requested].append(len(pass_ids))
    return out

def bin_pass_counts_by_qc(pass_func, coded_only, nr = None):
    base = coded_only_df() if coded_only else df
    f = base if nr is None else base[(base['num_requested'] == nr)]
    f_runs = f.groupby(['num_requested', 'hashed_worker_id', 'question_code'])
    
    out = defaultdict(list)
    
    for (num_requested, wid, qc), run in f_runs:
        pass_ids = pass_func(run)
        out[qc].append(len(pass_ids))
    return out
```

# Run Measures

Measures on runs that would have no meaning on a specific response

```python
def count_clusters(f):
    return len(set(f['fixed_cluster_number']))

def count_mix_ratio(run):
    """
    Returns #outmixes for each inmix in the run
    """
    roots = set()
    leaves = set()
    
    for i in run.index:
        j = i
        
        while(run['distance_fil'][j] > 0):
            j = j - run['distance_fil'][j]
            if not j in run.index:
                j = -1
                break
        if i != j and j != -1:
            roots.add(j)
            leaves.add(i)
        
        j = i
        while(run['distance_mike'][j] > 0):
            j = j - run['distance_mike'][j]
            if not j in run.index: # handle bugs where distances go to other runs or we have rows missing codes completely
                j = -1
                break
        if i != j and j != -1:
            roots.add(j)
            leaves.add(i)
    
    if len(leaves) == 0:
        return 0
    
    return float(len(leaves))/len(roots)
    
def counts_by_qc(count_func, coded_only, nr = None):
    base = coded_only_df() if coded_only else df
    f = base if nr is None else base[(base['num_requested'] == nr)]
    f_runs = f.groupby(['num_requested', 'hashed_worker_id', 'question_code'])
    
    out = defaultdict(list)
    
    for (num_requested, wid, qc), run in f_runs:
        out[qc].append(count_func(run))
    return out

def counts_by_nr(count_func, coded_only, qc = None):
    base = coded_only_df() if coded_only else df
    f = base if qc is None else base[(base['question_code'] == qc)]
    f_runs = f.groupby(['num_requested', 'hashed_worker_id', 'question_code'])
    
    out = defaultdict(list)
    
    for (num_requested, wid, qc), run in f_runs:
        out[num_requested].append(count_func(run))
    return out
```

# Cluster analysis

We are looking to answer the following questions:

+ number of clusters
+ distributions of number of items in each cluster
+ summary stats of the above normalized to number of ideas collected
+ how many original ideas people are getting, how many categories they are getting
+ individual contributions vis-a-vis the clusters: how many ideas to people have that are in other clusters? How big are the clusters? If someone is generating ideas found in big clusters, this person is generating unoriginal ideas

## Number of Clusters

```python
def num_clusters(question_code):
    sub_df = h_clustered[h_clustered['question_code'] == question_code]
    return len(set(sub_df['h_cluster']))

for qc in question_codes:
    print "Question code %s" % qc
    print "\tNumber of ideas:", len(df[df['question_code'] == qc])
    print "\tNumber of clustered ideas:", len(h_clustered[(h_clustered['question_code'] == qc)])
    
    if len(h_clustered[(h_clustered['question_code'] == qc)]) == 0:
           continue
    
    print '\tNumber of "low clusters" for question code', qc, ':', num_clusters(qc)
    thing = len(h_clustered[h_clustered['question_code'] == qc])/float(num_clusters(qc))
    print "\tresponses per cluster", thing
    
```

## Cluster size frequencies for h_clusters (low)

```python
for qc in question_codes:
    sub_df = h_clustered[(h_clustered['question_code'] == qc)]
    if len(sub_df) == 0:
        continue
    
    cluster_sizes = sub_df['h_cluster'].value_counts()
    cluster_sizes = [x for x in cluster_sizes if x > 0] # value_counts look for values that aren't there within the range
    bins = max(cluster_sizes) - 1
    
    hist(cluster_sizes, alpha=0.4, bins=bins, color=question_colors[qc])
    title("cluster sizes for h_clusters (low, %s)" % qc)
    xlabel("cluster size")
    ylabel("# of clusters at size")
    show()
```

## Cluster Size Frequencies for h_clusters (high)

```python
for qc in question_codes:
    sub_df = h_clustered[(h_clustered['question_code'] == qc)]
    if len(sub_df) == 0:
        continue
    
    cluster_sizes = sub_df['h_cluster_high'].value_counts()
    cluster_sizes = [x for x in cluster_sizes if x > 0] # value_counts look for values that aren't there within the range
    bins = max(cluster_sizes) - 1
    
    hist(cluster_sizes, alpha=0.4, bins=bins, color=question_colors[qc])
    title("cluster sizes for h_clusters (high, %s)" % qc)
    xlabel("cluster size")
    ylabel("# of clusters at size")
    show()
```
