```python
import csv
import pandas as pd
from scipy.stats.stats import pearsonr

responses = []

with open('/processed_data/strategy_coded_runs.csv', 'r') as f:
    csvr = csv.reader(f, delimiter=',')
    
    run_id = 0
    index = 0
    last = False
    
    for i, row in enumerate(csvr):
        if i == 0:
            continue
        
        if all([len(val) == 0 for val in row]):
            if last == False:
                run_id += 1
            last = True
            index = 0
        else:
            last = False
            
            nr = run_id / 5
            if nr == 0:
                nr = 5
            elif nr == 1:
                nr = 10
            elif nr == 2:
                nr = 25
            elif nr == 3:
                nr = 50
            elif nr == 4:
                nr = 75
            elif nr == 5:
                nr = 100
            
            responses.append((run_id, row[0], row[1], row[2], row[3], row[4],
                              ','.join([i for i in row[5:] if len(i) > 0]), index, nr))
            index += 1

df = pd.DataFrame({'run_id': pd.Series([r[0] for r in responses]),
                   'scoping': pd.Categorical.from_array([r[1] for r in responses]),
                   'riff_type': pd.Categorical.from_array([r[2] for r in responses]),
                   'pair': pd.Categorical.from_array([r[3] for r in responses]),
                   'reach_back': pd.Categorical.from_array([r[4] for r in responses]),
                   'partial': pd.Categorical.from_array([r[5] for r in responses]),
                   'answer': pd.Series([r[6] for r in responses]),
                   'answer_index': pd.Series(r[7] for r in responses),
                   'num_requested': pd.Series(r[8] for r in responses)
                   })

stat_cols = ['scoping', 'riff_type', 'pair', 'reach_back', 'partial']
```

```python
nice_names = {
              'scoping': {'I': 'inward scoping',
                          'o': 'outward scoping' },
              'riff_type': {'h': 'hold riffing',
                            'c': 'continuation riffing',
                            'r': 'repeat riffing',
                            'g': 'generalization riffing'},
              'pair': {'1': 'pairing'},
              'reach_back': {'1': 'reach back'},
              'partial': {'s': 'scoping without solution',
                          'pp': 'passing the problem',
                          'pg': 'plan without goal',
                          'gp': 'goal without plan'}
              }
```

```python
rids = list(set(df['run_id']))
run_metrics = {'run_id': pd.Series(rids)}

for c in stat_cols:
    
    vals = set(df[c])
    
    for v in vals:
        c_v_counts = []
        for rid in rids:
            c_v_counts.append(len(df[(df['run_id'] == rid) & (df[c] == v)]))
        
        run_metrics['%s_%s' % (c, v)] = pd.Series(c_v_counts)
        
rdf = pd.DataFrame(run_metrics)
```

```python
rs = []

for i, c1 in enumerate(rdf.columns[:-1]):
    if c1 == 'run_id' or c1[-1] == '_':
        continue
    for c2 in rdf.columns[i+1:]:
        if c2 == 'run_id'  or c2[-1] == '_':
            continue
        
        pr = pearsonr(rdf[c1], rdf[c2])
        rs.append((" ".join([c1, c2, 'pearsonr', str(pr)]), pr[0]))
        
print "\n".join([s[0] for s in sorted(rs, key=lambda x: abs(x[1]))])
```
