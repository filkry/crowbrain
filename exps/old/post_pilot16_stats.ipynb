```python
base_data_dirs = ['/processed_data/pilot16',
                  '/processed_data/pilot14',
                  '/processed_data/pilot13',
                  '/processed_data/pilot12',
                  '/processed_data/pilot11',]

manual_csvs = map(lambda x: '/processed_data/pilot16_metrics/' + x + '-scores.csv',
                         ['fil', 'mike'])

def metrics_folder(x):
    return '/processed_data/pilot16_metrics/' + x

idea_cluster_csvs = {qc: metrics_folder("_%s.csv" % qc) for qc in \
                     ['charity', 'iPod', 'forgot_name', 'turk']}
cluster_tree_csvs = {qc: metrics_folder("_%s_clusters.csv" % qc) for qc in \
                     ['charity', 'iPod', 'forgot_name', 'turk']}

nwf = lambda x: '/processed_data/pilot14_metrics/' + x + '-newick.csv'
output_newick = dict();
for key in ['charity', 'iPod', 'turk', 'forgot_name']:
    output_newick[key] = nwf(key)
    
output_csv = '/processed_data/pilot16_metrics/complete.csv'

# Read in a bunch of base data

def read_files(fils):
    rows = []
    for f in fils:
        rows = rows + read_file(f)
    return rows
     

def read_base_data(dirs):
    rows = []
    for d in dirs:
      for f in os.listdir(d):
        full_name = os.path.join(d,f)
        if os.path.isdir(full_name):
            print "I deleted the code here"
        else:
          if f == 'answers.csv':
            rows = rows + read_file(full_name)
    return rows

def read_file(f):
    with open(f) as fin:
        l = fin.readline()
        # If we see the follow, it's an early run we can ignore
        if 'Number of Answers' in l:
            return
        fin.seek(0)
        sep = '|'
        # Guess separator character
        if l.count(',') > l.count('|'):
            sep = ','
            
        # Just cache for now
        rows = []
        for row in csv.reader(fin, delimiter=sep):
            if 'hashed_worker_id' in row or 'worker_id' in row or 'cluster_parent' in row:
                continue
            rows.append(row + [f])
            
        return rows

def clean_missing(v):
    return 0 if v == '' or v == 'missing' else v

def series_from_row(rows, index, t):      
    if t == object:
        return pd.Series([row[index] for row in rows], dtype=t)
    elif t == datetime64:
        return pd.Series([np.datetime64(row[index]) for row in rows], dtype=t)
    else:
        s = pd.Series([clean_missing(row[index]) for row in rows], dtype=t)
        return s

merge_column_names = ['hashed_worker_id', 'question_code', 'answer_num', 'num_requested']
    
# Read in the main processed data files
all_rows = read_base_data(base_data_dirs)
num_responses = len(all_rows)

print "num base responses:", num_responses

bad_times = [(0 if (r[9] == 'missing' or r[10] == 'missing' or \
                    int(r[9]) <=0 or int(r[10]) <= 0) else 1) for r in all_rows]

df_base = pd.DataFrame({'hashed_worker_id': pd.Series([row[0] for row in all_rows], dtype=object),
            'question_code': pd.Series([row[2] for row in all_rows], dtype=object),
            'num_requested': pd.Series([row[5] for row in all_rows], dtype=uint8),
            'answer_num': pd.Series([row[6] for row in all_rows], dtype=uint8),
            'answer': pd.Series([row[7] for row in all_rows], dtype=object),
            'answer_word_count': pd.Series([row[8] for row in all_rows], dtype=uint32),
            'submit_datetime': pd.Series(pd.to_datetime([row[12] for row in all_rows])),
            'accept_datetime': pd.Series(pd.to_datetime([row[13] for row in all_rows])),
            'start_time': series_from_row(all_rows, 9, uint64),
            'end_time': series_from_row(all_rows, 10, uint64),
            'time_valid': pd.Series(bad_times, dtype=uint8),
            'batch_file': series_from_row(all_rows, 14, object),
})

df = df_base.copy()

# Read in heirarchical clusters
for key in idea_cluster_csvs:
    rows = read_file(idea_cluster_csvs[key])
    
    if len(rows) == 0:
        continue

    df = pd.merge(df,
                  pd.DataFrame({'hashed_worker_id': series_from_row(rows, 5, object),
                      'question_code': series_from_row(rows, 0, object),
                      'num_requested': series_from_row(rows, 7, uint8),
                      'answer_num': series_from_row(rows, 4, uint8),
                      'h_cluster': series_from_row(rows, 2, uint64),
                      'h_cluster_valid': pd.Series([1 for row in rows], dtype=uint8),}),
                  'outer', merge_column_names)
    
    

# Read in manual codes

def read_manual_csv(name):
    index = 0 if name == 'fil' else 1
    rows = read_file(manual_csvs[index])
    
    df = pd.DataFrame({'hashed_worker_id': series_from_row(rows, 0, object),
            'question_code': series_from_row(rows, 2, object),
            'num_requested': series_from_row(rows, 5, uint8),
            'answer_num': series_from_row(rows, 6, uint8),
            ('originality_%s' % name): series_from_row(rows, 12, uint8),
            ('utility_%s' % name): series_from_row(rows, 13, uint8),
            ('realistic_%s' % name): series_from_row(rows, 14, uint8),
            ('underdefined_%s' % name): series_from_row(rows, 15, uint8),
            ('distance_%s' % name): series_from_row(rows, 16, uint8),
            ('%s_valid' % name): pd.Series([1 for row in rows], dtype=uint8),
    })
    return df

df = pd.merge(df, read_manual_csv('mike'), 'outer', merge_column_names)
df = pd.merge(df, read_manual_csv('fil'), 'outer', merge_column_names)

print "Dropping unlimited condition"
df = df[~(df['num_requested'] == 0)]

print "num merged responses:", len(df)

print "Sorting by response order"
df = df.sort(['submit_datetime'])
```

# Generate cluster trees and tree metrics

Note: all of this is for convenience, and assumes the tree structure is static!

+ cluster o_score
+ cluster simple_originality

```python
def dumb_strip(s):
    return ''.join([c for c in s if c in "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz1234567890 "])
```

```python
clustered_df = df[(df['h_cluster_valid'] > 0)]
clustered_qcs = list(set(df[(df['h_cluster_valid'] > 0)]['question_code']))

cluster_forests = {qc: nx.DiGraph() for qc in clustered_qcs}

def all_nodes_under(digraph, node):
    ret = [node]
    for s in digraph.successors(node):
        ret += all_nodes_under(digraph, s)
    return ret

def num_ideas_in(adf, clusters):
    return sum(len(adf[(adf['h_cluster'] == c)]) for c in clusters)

def gen_depth(forest, node):
    parents = forest.predecessors(node)
    if parents:
        p = parents[0]
        if 'depth' not in forest.node[p]:
            gen_depth(forest, p)
        forest.node[node]['depth'] = forest.node[p]['depth'] + 1
    else:
        forest.node[node]['depth'] = 0
    
    smd = f.graph['subtree_max_depth']
    root = f.node[node]['subtree_root']
    smd[root] = max(smd[root], forest.node[node]['depth'])
                                       
for qc in clustered_qcs:
    f = cluster_forests[qc]
    f.graph['subtree_roots'] = set()
    f.graph['subtree_max_depth'] = defaultdict(int)
    sub_df = clustered_df[(clustered_df['question_code'] == qc)]
    
    rows = read_file(cluster_tree_csvs[qc])
    for qc, child, parent, label, fn in rows:
        c = int(child)
        p = int(parent)
        if parent != '':
            f.add_edge(p, c)
        else:
            f.add_node(c)
        f.node[c]['label'] = dumb_strip(label) if len(label) > 0 else None
            
    # delete the single root node
    root = nx.topological_sort(f)[0]
    f.remove_nodes_from([root])
    
    # Subtree root
    for node in f.nodes():
        cur = node
        while(len(f.predecessors(cur)) > 0):
            cur = f.predecessors(cur)[0]
        f.node[node]['subtree_root'] = cur
        f.graph['subtree_roots'].add(cur)
            
    # Cache the set of all nodes under each node
    for node in f.nodes():
        f.node[node]['all_under'] = all_nodes_under(f, node)

    # Cache the number of ideas under each node
    f.graph['max_num_ideas_under'] = 0
    for node in f.nodes():
        f.node[node]['num_ideas_under'] = num_ideas_in(sub_df, f.node[node]['all_under'])
        f.graph['max_num_ideas_under'] = max(f.graph['max_num_ideas_under'],
                                             f.node[node]['num_ideas_under'])
        
    # compute probability for each cluster
    for node in f.nodes():
        f.node[node]['probability'] = float(f.node[node]['num_ideas_under']) / len(sub_df)
    
    # simple originality score
    for node in f.nodes():
        s = 1 - float(f.node[node]['num_ideas_under']) / f.graph['max_num_ideas_under']
        f.node[node]['simple_originality'] = s
        
    # depth
    for node in f.nodes():
        if 'depth' not in f.node[node]:
            gen_depth(f, node)
            
    # height
    for n in f.nodes():
        smd = f.graph['subtree_max_depth']
        root = f.node[n]['subtree_root']
        f.node[n]['height'] = smd[root] - f.node[n]['depth']
        
    # remix_of
    for n in f.nodes():
        targets = set([n])
        
        # all parents
        cur = n
        while len(f.predecessors(cur)) > 0:
            cur = f.predecessors(cur)[0]
            targets.add(cur)
        
        # all siblings
        for p in f.predecessors(n):
            for s in f.successors(p):
                targets.add(s)
        
        # all children
        targets = targets.union(set(f.node[n]["all_under"]))
        
        f.node[n]['remix_of'] = targets
        
```

## Put cluster metrics in their own dataframe

```python
ids = []
labels = []
is_roots = []
is_leafs = []
roots = []
num_unders = []
probabilitys = []
simple_originalitys = []
depths = []
heights = []
qcs = []
num_nodes_under = []
num_childrens = []

for qc in clustered_qcs:
    f = cluster_forests[qc]
    for n in f.nodes():
        nd = f.node[n]
        
        ids.append(n)
        labels.append(nd['label'])
        is_roots.append(n == nd['subtree_root'])
        is_leafs.append(len(f.successors(n)) == 0)
        num_childrens.append(len(f.successors(n)))
        roots.append(nd['subtree_root'])
        num_unders.append(nd['num_ideas_under'])
        probabilitys.append(nd['probability'])
        simple_originalitys.append(nd['simple_originality'])
        depths.append(nd['depth'])
        heights.append(nd['height'])
        qcs.append(qc)
        num_nodes_under.append(len(nd['all_under']))

clusters_df = pd.DataFrame({
        'h_cluster': pd.Series(ids, dtype=uint64),
        'label': pd.Series(labels, dtype=object),
        'is_root': pd.Series(is_roots, dtype=uint8),
        'is_leaf': pd.Series(is_leafs, dtype=uint8),
        'subtree_root': pd.Series(roots, dtype=uint64),
        'num_ideas_under': pd.Series(num_unders, dtype=uint64),
        'probability': pd.Series(probabilitys, dtype=float64),
        'simple_originality_score': pd.Series(simple_originalitys, dtype=float64),
        'depth_in_subtree': pd.Series(depths, dtype=uint32),
        'height_in_subtree': pd.Series(heights, dtype=uint32),
        'question_code': pd.Series(qcs, dtype=object),
        'num_nodes_under': pd.Series(num_nodes_under, dtype=uint64),
        'num_children': pd.Series(num_childrens, dtype=uint64),
    })

clusters_df['o_score'] = 1 - clusters_df['probability']
```

### Outmix/inmix

```python
def compute_mixing():
    dist = pd.Series([None for i in df.index], index=df.index)
    dist_im = pd.Series([None for i in df.index], index=df.index)
    im = pd.Series([0 for i in df.index], index=df.index)
    om = pd.Series([0 for i in df.index], index=df.index)
    last_sim = pd.Series([None for i in df.index], index=df.index)
    
    for (nr, wid, qc), run in df.groupby(['num_requested', 'hashed_worker_id', 'question_code']):
        if qc not in question_codes:
            continue
        for ii, i in enumerate(run.index):
            if ii == 0:
                continue
            for jj, j in reversed(list(enumerate(run.index[0:ii]))):
                j_clus = run['h_cluster'][j]
                hcm_nc = cluster_forests[qc].node[j_clus]['remix_of']
                if run['h_cluster'][i] in hcm_nc:
                    last_sim[i] = j
                    dist[i] = ii - jj
                    
                    if om[j] > 0:
                        dist_im[i] = dist[i] + dist_im[j]
                    else:
                        im[j] = 1
                        dist_im[i] = dist[i]
                    
                    om[i] = 1
                    assert dist[i] > 0
                    break
    return dist, im, om, dist_im, last_sim

dist, im, om, dist_im, last_sim = compute_mixing()

df['distance_from_similar'] = dist
df['is_inmix'] = im
df['is_outmix'] = om
df['distance_from_inmix'] = dist_im
df['previous_similar'] = last_sim
```

```python
def run_count(run, pass_func):
    count = 0
    for i in run.iterrows():
        row = i[1]
        if pass_func(row):
            count += 1
    return count

def run_mean(run, value_func):
    t = 0.0
    for i in run.iterrows():
        row = i[1]
        t += value_func(row)
    return t / len(run)

runs = df.groupby(['num_requested', 'hashed_worker_id', 'question_code', 'submit_datetime'])

wids = pd.Series([wid for ((nr, wid, qc, sdt), run) in runs], dtype=object)
qcs = pd.Series([qc for ((nr, wid, qc, sdt), run) in runs], dtype=object)
nrs = pd.Series([nr for ((nr, wid, qc, sdt), run) in runs], dtype=uint16)
nrc = pd.Series([len(run) for ((nr, wid, qc, sdt), run) in runs], dtype=uint16)
cpr = pd.Series([n for n in nrc], dtype=float64) / nrs

for i in nrs.index:
    assert(nrs[i] >= nrc[i])

sdts = pd.Series(pd.to_datetime([sdt for (nr, wid, qc, sdt), run in runs]))

rwq = pd.Series([run['repeat_wid_qc'].iloc[0] for (nr, wid, qc, sdt), run in runs], dtype=uint8)

mwc_val = lambda x: x['answer_word_count']
mwc = pd.Series([run_mean(run, mwc_val) for ((nr, wid, qc, sdt), run) in runs],
                dtype=float64)

nmo_pass = lambda x: x['originality_mike'] > 0 and \
                     x['originality_fil'] > 0 and \
                     x['distance_mike'] == 0 and \
                     x['distance_fil'] == 0
nmo = pd.Series([run_count(run, nmo_pass) for ((nr, wid, qc, sdt), run) in runs],
                dtype=uint16)

nmo_norm = pd.Series([float(n) / nrc[i] for (i, n) in enumerate(nmo)],
                 dtype=float64)

nmu_pass = lambda x: x['utility_mike'] == 0 and \
                     x['utility_fil'] == 0
nmu = pd.Series([run_count(run, nmu_pass) for ((nr, wid, qc, sdt), run) in runs],
                dtype=uint16)

nmu_norm = pd.Series([float(n) / nrc[i] for (i, n) in enumerate(nmu)],
                     dtype=float64)

# TODO: these measures can't be trusted
nu = pd.Series([len(set(run['h_cluster'])) for ((nr, wid, qc, sdt), run) in runs],
                dtype=uint16)

pu = pd.Series([float(n) / nrc[i] for (i, n) in enumerate(nu)],
                     dtype=float64)

# Realistic if neither of us voted it unrealistic
nmr_pass = lambda x: x['realistic_mike'] == 0 and \
                     x['realistic_fil'] == 0
nmr = pd.Series([run_count(run, nmr_pass) for ((nr, wid, qc, sdt), run) in runs],
                dtype=uint16)

nmr_norm = pd.Series([float(n) / nrc[i] for (i, n) in enumerate(nmr)],
                     dtype=float64)

nmun_pass = lambda x: x['underdefined_mike'] > 0 and \
                     x['underdefined_fil'] > 0
nmun = pd.Series([run_count(run, nmun_pass) for ((nr, wid, qc, sdt), run) in runs],
                dtype=uint16)
```

# Graphing Functions

These were getting long and generic, so put them where they can be safely ignored

```python
def init_multifig(num_conditions, fignum = 1):
    rows = math.ceil(num_conditions / float(2))
    return rows, figure(fignum, figsize=(14, rows * 5.5))

def hists(conditions, plot_title, x_title, y_title, fignum = 1, thin_bins = False):
    rc, fig = init_multifig(len(conditions), fignum)
    
    all_hist_counts = []
    max_x = 0
    for (con, color, data) in conditions:
        max_x = max(max_x, max(data))
        if type(data) is pd.Series:
            for d in data.value_counts():
                if d != 0:
                    all_hist_counts.append(d)
        else:
            for d in np.bincount(data):
                all_hist_counts.append(d)
    ymax = max(all_hist_counts)
    
    for i, (con, color, data) in enumerate(conditions):
        subplot(rc, 2, i)
        
        if thin_bins:
            hist(data, alpha=0.4, color=color, bins=max_x)
        else:
            hist(data, alpha=0.4, color=color)
        title(plot_title + ' (' + str(con) + ')')
        xlabel(x_title)
        ylabel(y_title)
        ylim(0, ymax)

def hists_cross_nr(base_df, field, plot_title, x_title, y_title, fignum = 1):
    conditions = []
    
    for nr in number_conditions:
        nr_df = base_df[(base_df['num_requested']) == nr]
        
        if len(nr_df) == 0:
            continue
        
        conditions.append((nr, question_colors[None], nr_df[field]))
        
    hists(conditions, plot_title, x_title, y_title, fignum)

def bar_mean_err(conditions, plot_title, x_title, x_labels, y_title, fignum = 1):
    rc, fig = init_multifig(len(conditions), fignum)
    
    firstax = None
    
    ymin = None
    for c, c2, bar_datas in conditions:
        for bd in bar_datas:
            if len(bd) > 0:
                if not ymin is None:
                    ymin = min(ymin, min(bd))
                else:
                    ymin = min(bd)
    
    for i, (con, color, bar_datas) in enumerate(conditions):
        assert(len(x_labels) == len(bar_datas))
        
        means = [np.mean(bd) for bd in bar_datas]
        stes = [stats.sem(bd, ddof=1) for bd in bar_datas]

        if firstax:
            ax = subplot(rc, 2, i, sharey=firstax)
        else:
            firstax = subplot(rc, 2, i)
            
        bar(range(len(means)), means, yerr=stes, alpha=0.4, color=color)
        xticks(range(len(means)), x_labels)
        
        title(plot_title + ' (' + str(con) + ')')
        xlabel(x_title)
        ylabel(y_title)
        ylim(ymin=ymin)
        
def bar_mean_err_cross_qc(base_df, field, plot_title, y_title, fignum = 1):
    conditions = []
    
    bar_datas = []
    for nr in number_conditions:
        nr_df = base_df[(base_df['num_requested']) == nr]
        bar_datas.append(nr_df[field])
    conditions.append(("all", question_colors[None], bar_datas))
    
    for qc in question_codes:
        qc_df = base_df[(base_df['question_code'] == qc)]
        bar_datas = []
        for nr in number_conditions:
            qc_nr_df = qc_df[(qc_df['num_requested']) == nr]
            bar_datas.append(qc_nr_df[field])
        conditions.append((qc, question_colors[qc], bar_datas))
        
    bar_mean_err(conditions, plot_title, '# responses requested',
                 number_conditions, y_title, fignum)
    
def bar_mean_err_cross_nr(base_df, field, plot_title, y_title, fignum = 1):
    conditions = []
    
    bar_datas = []
    for qc in question_codes:
        qc_df = base_df[(base_df['question_code']) == qc]
        bar_datas.append(qc_df[field])
    conditions.append(("all", [question_colors[qc] for qc in question_codes], bar_datas))
    
    for nr in number_conditions:
        nr_df = base_df[(base_df['num_requested'] == nr)]
        if len(nr_df) == 0:
            continue
        bar_datas = []
        for qc in question_codes:
            qc_nr_df = nr_df[(nr_df['question_code']) == qc]
            bar_datas.append(qc_nr_df[field])
        conditions.append((nr, [question_colors[qc] for qc in question_codes], bar_datas))
        
    bar_mean_err(conditions, plot_title, '# responses requested',
                 question_codes, y_title, fignum)


def count_over_responses_cross_qc(base_df, count_func, xlab, ylab, t):
    rows, fig = init_multifig(len(question_codes))

    for i, qc in enumerate(question_codes):
        sub_df = base_df[(base_df['question_code'] == qc)]
        subplot(rows, 2, i + 1)
        xlabel(xlab)
        ylabel(ylab)
        title('%s (%s)' % (t, qc))
        
        min_cond_size = min(len(sub_df[(sub_df['num_requested'] == nr)]) for nr in number_conditions)
        
        for nr in number_conditions:
            ssub_df = sub_df[(sub_df['num_requested'] == nr)]
            ssub_df = ssub_df.sort('submit_datetime')
            ssub_df = ssub_df.iloc[:min_cond_size]
            
            xs, ys = zip(*count_func(ssub_df))
            
            plot(xs, ys, label=str(nr))
        
        legend(loc='upper left')
```
