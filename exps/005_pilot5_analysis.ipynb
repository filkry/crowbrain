# Import Data

```python
import sqlite3
conn = sqlite3.connect(':memory:')

conn.execute("CREATE TABLE IF NOT EXISTS data (worker_id TEXT, question TEXT, total_answers INTEGER, \
    answer_num INTEGER, answer TEXT, start_time INTEGER, end_time INTEGER, coded_answer INT)")

with open('/processed_data/pilot5/answers.csv') as ans:
    reader = csv.reader(ans)
    for i, row in enumerate(reader):
        if i == 0:
            print row
            continue
            
        # Exclude users with broken questions
        if not "thumb" in row[1] and not "weight" in row[1] and not "mop" in row[1]:
            continue
            
        # Exclude null answers (need to exclude users entirely down the line)
        if row[4] == '':
            continue
            
        conn.execute("INSERT INTO data VALUES (?, ?, ?, ?, ?, ?, ?, ?)", row)
        
        
def np_select(sql):
    l = []
    
    for row in conn.execute(sql):
        l.append(row)
    
    return np.array(l)

# quick test
# np_select("SELECT question, answer_num, answer FROM data WHERE answer_num=0")
        
```

# Various word count analysis

## Word counts overall

```python
def overall_word_counts(question_sub, total_answers):
    gross_sql = "SELECT answer FROM data WHERE question LIKE '%" + question_sub + ("%%' AND total_answers=%s ORDER BY worker_id" % (total_answers))
    ar = np_select(gross_sql)
    
    wcs = []
    
    for i in range(len(ar) / total_answers):
        start = i * total_answers
        end = (i+1) * total_answers
        answers = ar[start:end]
        wcs.append(sum([len(a[0]) for a in answers]))
    
    return wcs

def wordcount_plot(question_sub, plot_pos):
    total_answers = [5, 20, 50]
    word_counts = np.array([overall_word_counts(question_sub, 5),
                            overall_word_counts(question_sub, 20),
                            overall_word_counts(question_sub, 50)])
    
    means = [np.mean(word_counts[i]) for i in range(len(word_counts))]
    stes = [sem(word_counts[i], ddof=1) for i in range(len(word_counts))]
    
    subplot(plot_pos)
    errorbar(total_answers, means, yerr=stes, marker='o', label=question_sub)
    xlim([0, max(total_answers) + 5])
    title("word counts for %s" % question_sub)
    ylabel("word count")
    xlabel("total responses solicited")
    
    
fig = figure(1, figsize=(14,5))
wordcount_plot("thumb", 111)
wordcount_plot("weight", 111)
wordcount_plot("mop", 111)

legend()
```

## Word counts per idea

```python
def pi_word_counts(question_sub, total_answers):
    gross_sql = "SELECT answer FROM data WHERE question LIKE '%" + question_sub + ("%%' AND total_answers=%s ORDER BY worker_id" % (total_answers))
    ar = np_select(gross_sql)
    
    wcs = []
    
    for a in ar:
        wcs.append(len(a[0]))
    
    return wcs

def pi_wordcount_plot(question_sub, plot_pos):
    total_answers = [5, 20, 50]
    word_counts = np.array([pi_word_counts(question_sub, 5),
                            pi_word_counts(question_sub, 20),
                            pi_word_counts(question_sub, 50)])
    
    means = [np.mean(word_counts[i]) for i in range(len(word_counts))]
    stes = [sem(word_counts[i], ddof=1) for i in range(len(word_counts))]
    
    subplot(plot_pos)
    errorbar(total_answers, means, yerr=stes, marker='o', label=question_sub)
    xlim([0, max(total_answers) + 5])
    title("per idea word counts for %s" % question_sub)
    ylabel("word count")
    xlabel("total responses solicited")
    
    
fig = figure(1, figsize=(14,7))
pi_wordcount_plot("thumb", 111)
pi_wordcount_plot("weight", 111)
pi_wordcount_plot("mop", 111)

legend()
```

## Word Counts as function of response number

```python
def per_question_word_counts(question_sub, total_answers):
    wc_means_by_answer_num = []
    wc_stes_by_answer_num = []
    
    for i in range(total_answers):
        gross_sql = "SELECT answer FROM data WHERE question LIKE '%" + question_sub + ("%%' AND total_answers=%s AND answer_num=%d" % (total_answers, i))
        ar = np_select(gross_sql)

        wcs = [len(a[0]) for a in ar]
        
        wc_means_by_answer_num.append(np.mean(wcs))
        wc_stes_by_answer_num.append(sem(wcs, ddof=1))
        
    return wc_means_by_answer_num, wc_stes_by_answer_num

def per_question_wordcount_plot(question_sub, total_answers, plot_pos): 
    means, stes = per_question_word_counts(question_sub, total_answers)
    
    subplot(plot_pos)
    errorbar(range(total_answers), means, yerr=stes, marker='o', label=question_sub)
    xlim([-1, total_answers + 1])
    title("word counts for requesting %d" % (total_answers))
    ylabel("word count")
    xlabel("answer number")

fig = figure(1, figsize=(14,14))

for count, ax in [(5, 311), (20, 312), (50, 313)]:
    per_question_wordcount_plot("thumb", count, ax)
    per_question_wordcount_plot("weight", count, ax)
    per_question_wordcount_plot("mop", count, ax)
    
    legend()
```

# Time spent analysis

## Time spent as function of response number

```python
def per_reponse_time_spent(question_sub, total_answers):
    time_means_pr = []
    time_stes_pr = []
    
    for i in range(total_answers):
        gross_sql = "SELECT end_time, start_time FROM data WHERE question LIKE '%" + question_sub + ("%%' AND total_answers=%s AND answer_num=%d" % (total_answers, i))
        ar = np_select(gross_sql)
        
        time = [int(a[0]) - int(a[1]) for a in ar]
        
        time_means_pr.append(np.mean(time))
        time_stes_pr.append(sem(time, ddof=1))
        
    return time_means_pr, time_stes_pr

def per_response_time_spent_plot(question_sub, total_answers, plot_pos): 
    means, stes = per_reponse_time_spent(question_sub, total_answers)
    
    subplot(plot_pos)
    errorbar(range(total_answers), means, yerr=stes, marker='o', label=question_sub)
    xlim([-1, total_answers + 1])
    title("time spent on answer for requesting %d" % (total_answers))
    ylabel("time spent (need unit)")
    xlabel("answer number")

fig = figure(1, figsize=(14,14))

for count, ax in [(5, 311), (20, 312), (50, 313)]:
    per_response_time_spent_plot("thumb", count, ax)
    per_response_time_spent_plot("weight", count, ax)
    per_response_time_spent_plot("mop", count, ax)
    
    legend()
```
